### csv to html

Import dependencies

```python
import pandas as pd
import os
```

Import CSVs

```python
city_file = os.path.join("Resources", "clean_city_data.csv")

stats_file = os.path.join("Resources", "summary_stats.csv")
```

```python
city_df = pd.read_csv(city_file)
city_df.head()

```

```python
stats_df = pd.read_csv(stats_file)
stats_df.head()
```

Convert to HTML

```python
city_html = city_df.to_html()
print(city_html)
```

```python
stats_html = stats_df.to_html()
print(stats_html)
```
